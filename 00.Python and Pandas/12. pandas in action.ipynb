### clean salary

- bottom and floor
- capital k
- '12k以上'
- '2k以下'

````python
df.salary.unique()
````

````python
# split into two columns
salary_ = df.salary.str.lower().str.split('-')
salary_bottom = salary_.str.get(0).str.replace('以上', '')
salary_top = salary_.str.get(1)
````

````python
salary_top.unique()
````

````python
salary_bottom.unique()
````

````python
# clean '以下'
salary_top_mask = \
    salary_top.mask(salary_bottom.str.contains('以下'), 
                          salary_bottom)

salary_bottom_mask = \
    salary_bottom.mask(salary_bottom.str.contains('以下'), 
                          np.nan)
  
````

````python
salary_top_mask.unique()
````

````python
salary_bottom_mask.unique()
````

````python
salary_top_mask = salary_top_mask.str.replace('以下', '')
````

````python
salary_top_mask.unique()
````

````python
df_cl = df.assign(salary_top=salary_top_mask, 
                  salary_bottom=salary_bottom_mask)
````

````python
# '以上'
df_cl[df_cl.salary_top.isnull()][['salary_bottom', 'salary_top']]
````

````python
# '以下'
df_cl[df_cl.salary_bottom.isnull()][['salary_bottom', 'salary_top']]
````

````python
# fill bottom top
df_cl.salary_bottom.fillna(df_cl.salary_top, inplace=True)
df_cl.salary_top.fillna(df_cl.salary_bottom, inplace=True)
````

````python
df_cl2 = df_cl.assign(
    salary_bottom_int = df_cl.salary_bottom.str.replace('k', '').apply(int) * 1_000,
    salary_top_int = df_cl.salary_top.str.replace('k', '').apply(int) * 1_000,)
````

````python
df_cl2 = df_cl2.assign(
    avg_salary = ( df_cl2.salary_bottom_int + df_cl2.salary_top_int) / 2, )
````
