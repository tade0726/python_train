## Pandas vs SQL

- merge
- join
- pivotable
- change index

```python
import pandas as pd
import numpy as np
from pandasql import sqldf

from IPython.display import display
```

```python
first_time = False

if first_time:
    # loading data from pandas demo
    url = 'https://raw.github.com/pandas-dev/pandas/master/pandas/tests/data/tips.csv'
    tips = pd.read_csv(url)
    # dump as csv
    tips.to_csv('../data/tips.csv')
else:
    # loading 
    tips = pd.read_csv('../data/tips.csv')
```

## pandas style

```python
df_result1 = tips.groupby(['sex', 'smoker', 'day'])['day'].size()\
    .sort_index(axis=0, level=['sex', 'smoker'], ascending=[1, 1])\
    .rename('day_counts')\
    .reset_index()
```

## sql style

```python
pysqldf = lambda q: sqldf(q, globals())

# SQL
sql =\
"""
SELECT sex, smoker, day, COUNT(day) AS day_counts 
FROM tips
GROUP BY sex, smoker, day
ORDER BY sex, smoker ASC;
"""

df_result2 = pysqldf(sql)
```

```python
display(df_result1, df_result2)
```
