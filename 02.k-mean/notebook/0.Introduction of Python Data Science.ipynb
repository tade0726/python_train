## groupby

```python
tips.groupby(by=pd.cut(tips['size'], bins=[0, 1, 2, 4, 6]))['tip']\
    .agg({'Count': np.count_nonzero, 'Mean': np.mean, '50%': np.median, 'Sum': np.sum, })
```

```python
tips.groupby(tips.total_bill // 10 * 10)['tip']\
    .agg({'Count': np.count_nonzero, 'Mean': np.mean, '50%': np.median, 'Sum': np.sum, })
```

## pandas data transform

```python
#  1. pandas style
df_result_pd = tips.groupby(['sex', 'smoker', 'day'])['day'].size()\
    .sort_index(axis=0, level=['sex', 'smoker'], ascending=[1, 1])\
    .rename('txn_counts')\
    .reset_index()

# 2. SQL sytle
pysqldf = lambda q: sqldf(q, globals())

# SQL
sql =\
"""
SELECT sex, smoker, day, COUNT(day) AS txn_counts
FROM tips
GROUP BY sex, smoker, day
ORDER BY sex, smoker ASC;
"""
df_result_sql = pysqldf(sql)
```

```python
display(df_result_pd, df_result_sql)
```

### Broadcast
#### get the smoker percentage by day of week

```python
# get two frame
txn_by_day_smk = tips.groupby(by=['day', 'smoker']).size()
txn_by_day = tips.groupby(by=['day']).size()
# devide along the index
df_smoker_percentage = txn_by_day_smk / txn_by_day
df_smoker_percentage.rename('pct_smoker').reset_index()
```

```python
txn_by_day
```

```python
txn_by_day_smk
```

```python
txn_by_day_smk / txn_by_day
```

```python
# unstack
df_smoker_percentage.unstack(level=1)
```

```python
# in SQL style

# SQL
sql2 =\
"""
SELECT z.day, z.smoker, (z.two_counts  * 1.0 / z.one_counts) as pct_smoker 
FROM 
(SELECT *
    FROM (SELECT day, smoker, COUNT(*) AS two_counts 
            FROM tips
            GROUP BY day, smoker) as x
LEFT JOIN
(SELECT day, COUNT(*) AS one_counts 
        FROM tips 
        GROUP BY day) as y
USING (day)) as z;
"""
df_smoker_percentage_sql = pysqldf(sql2)
df_smoker_percentage_sql
```

```python
from sklearn import datasets
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline

# loading data
iris_data = sns.load_dataset("iris")
```

```python
# loading model
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
# for data visulization
from IPython.display import Image
import pydotplus 
```

```python
X = iris_data.drop('species', axis=1).values
y = iris_data['species'].values
feature_names = iris_data.columns[:-1]

# data split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# decision tree model
tree_cls = tree.DecisionTreeClassifier(criterion='entropy',
               max_leaf_nodes=10,
               # The minimum number of samples required to split an internal node
               min_samples_split=5,
               # The minimum number of samples required to be at a leaf node
               min_samples_leaf=5,)

# tree_cls = tree.DecisionTreeClassifier()
tree_train = tree_cls.fit(X_train, y_train)

# predict
y_pre_test = tree_train.predict(X_test)
y_pre_train = tree_train.predict(X_train)
```

```python
print(classification_report(y_train, y_pre_train))
```

```python
print(classification_report(y_test, y_pre_test))
```

```python
dot_data = tree.export_graphviz(tree_train, out_file=None,
                     feature_names=iris_data.columns[:-1],
                     filled=True, rounded=True,
                     special_characters=True,
                     node_ids=1,
                     proportion=1,)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
